# RL-lab-1 SOLUTION: The agent-environment interaction 

This is a sample solution to exercise 1. It includes methods for visualization and additional functionality to change the environment and the agent. Most code is imported from the Python files gridworld.py and agent.py which have to lie in the same directory as this notebook file.

There is also a more "lightweight"-version (in terms of lines of code) with less extra functionality that should let you focus more on the essential parts of the algorithm, instead of trying to understand all the weirdness of matplotlib :)

```python
import gridworld
import agent

import numpy as np
import time

%matplotlib inline
```

## The `play()` function

This function lets the agent interact with the environment for a specified number of periods and returns the cumulative reward per episode, which we then can visualize.

```python
def play(environment, agent, episodes=500, max_steps_per_episode=1000, learn=False, seconds_between_each_step=0,
        show_grid=False, show_policy=False, show_q_values=True, show_softmax=False, show_learning_curve=False):
    
    reward_per_episode = np.zeros(episodes)
    for episode in range(0, episodes):
        cumulative_reward = 0
        step = 0
        game_over = False
        while step < max_steps_per_episode and not game_over:
            time.sleep(seconds_between_each_step)
            if show_grid or show_learning_curve:
                environment.visualize(show_grid=show_grid, show_policy=show_policy, 
                                      show_softmax=show_softmax,
                                      show_learning_curve=show_learning_curve, 
                                      show_q_values=show_q_values, clear_the_output=True,
                                      episode=episode, reward_per_episode=reward_per_episode, 
                                      agent_q_table=agent.q_table)
                

            old_state = environment.agent_position
            action = agent.choose_action()
            reward, new_state = environment.make_step(action)
            if learn:
                agent.learn(old_state, reward, new_state, action)
            cumulative_reward += reward
            step += 1
            
            # Check whether agent is at terminal state. If yes: end episode; reset agent.
            if environment.is_terminal_state():
                time.sleep(seconds_between_each_step)
                if show_grid or show_learning_curve:
                    environment.visualize(show_grid=show_grid, show_policy=show_policy, 
                                          show_softmax=show_softmax,
                                          show_learning_curve=show_learning_curve, 
                                          show_q_values=show_q_values, clear_the_output=True,
                                          episode=episode, reward_per_episode=reward_per_episode, 
                                          agent_q_table=agent.q_table)
                environment.reset()
                game_over = True
                
                

        reward_per_episode[episode] = cumulative_reward
    return reward_per_episode
```

## Learning curve for a random agent

```python
np.random.seed(1)
environment = gridworld.Gridworld()
random_agent = agent.RandomAgent(environment)

# Do not visualize during play.
reward_per_episode = play(environment, random_agent, episodes=500)

# Learning curve.
environment.visualize(show_learning_curve=True, show_grid=False, show_policy=False, show_softmax=False,
                      episode=len(reward_per_episode), reward_per_episode=reward_per_episode)
```

## Visualization of policy and learning curve for a Q-learning agent
We now let the q-learning agent play and learn (without watching) and then visualize its policy afterwards.

```python
## Initialize environment and Q-learning agent.
np.random.seed(1)
environment = gridworld.Gridworld(load_images=True)
agentQ = agent.AgentQ(environment, policy="epsilon_greedy", alpha = 0.1, gamma = 1)

# Do not visualize during play.
reward_per_episode = play(environment, agentQ, episodes=500, learn=True)

# Plot policy and learning curve.
environment.visualize(show_grid=True, show_policy=True, show_learning_curve=True,
                      episode=len(reward_per_episode), reward_per_episode=reward_per_episode, 
                      agent_q_table=agentQ.q_table)
```

## Visualization of learning process for a Q-learning agent

```python
## Let agentQ play and watch while she is learning
np.random.seed(2)

# Define the envronment
environment = gridworld.Gridworld(load_images=True)

# Define the Q-learning agent
agentQ = agent.AgentQ(environment, policy="epsilon_greedy", alpha = 0.1, gamma = 1, epsilon=0.001)

reward_per_episode = play(environment, agentQ, episodes=500, learn=True,
                          show_grid=True, show_policy=False, 
                          show_q_values=True, show_softmax=False, 
                          show_learning_curve=True)
```
